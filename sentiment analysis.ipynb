```python
from transformers import pipeline

# Load the multilingual model and set up sentiment-analysis pipeline
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
sentiment_analyzer = pipeline("sentiment-analysis", model=model_name)
```

```python
import gradio as gr

# Define the function to analyze sentiment
def analyze_sentiment(text):
    # Run the text through the sentiment analysis pipeline
    result = sentiment_analyzer(text)
    label = result[0]['label']
    score = result[0]['score']
    return f"Sentiment: {label}, Confidence: {score:.2f}"

# Set up the Gradio interface
iface = gr.Interface(
    fn=analyze_sentiment,
    inputs="text",
    outputs="text",
    title="Multilingual Sentiment Analysis",
    description="Analyze sentiment in multiple languages!"
)
```

```python
iface.launch(share=True)
```

```python
test.strip()

test_list = test.split('.')
test_list = [sentence.strip() for sentence in test_list if sentence.strip()]
```

```python
total_confidence = 0
lowest_score = float('inf')
highest_score = float('-inf')
lowest_label = ""
highest_label = ""

for sentence in test_list:
    result = sentiment_analyzer(sentence)
    score = result[0]['score']
    label = result[0]['label']
    total_confidence += score

    if score < lowest_score:
        lowest_score = score
        lowest_label = label

    if score > highest_score:
        highest_score = score
        highest_label = label

average_confidence = total_confidence / len(test_list) if test_list else 0
```

```python

print("Average Confidence Score:", average_confidence)
print("Lowest Confidence Score:", lowest_score, "Label:", lowest_label)
print("Highest Confidence Score:", highest_score, "Label:", highest_label)
```
